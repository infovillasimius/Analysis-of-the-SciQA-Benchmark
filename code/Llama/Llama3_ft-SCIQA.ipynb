```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "Llama_3.1-8B_10_epochs",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
FastLanguageModel.for_inference(model)
```

```python
from datasets import load_dataset, Dataset
raw_dataset = load_dataset("orkg/SciQA", split = "test")
questions = []
for sample in raw_dataset:
    questions.append(sample["question"]["string"])
```

```python
!pip install tqdm
import json
from tqdm import tqdm
```

```python
def clean(st):
    st = st.replace("\n", " ")
    st = st.replace("?", " ?")
    st = st.replace("{", " { ")
    st = st.replace("}", " } ")
    st = st.replace("\\'", "'")

    while "  " in st:
        st = st.replace("  ", " ")
    return st

def write_json(file__name, content):
    with open(file__name, "w", encoding="utf-8") as text_file:
        print(json.dumps(content, indent=4), file=text_file)

sciqa_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

answers = []
for question in tqdm(questions):
    inputs = tokenizer(
    [
        sciqa_prompt.format(
            "Given an English text describing an information need, generate a SPARQL query to retrieve relevant data from the Open Research Knowledge Graph (ORKG)", # instruction
            question, # input
            "", # output - leave this blank for generation!
        )
    ], return_tensors = "pt").to("cuda")

    outputs = model.generate(**inputs,  use_cache = True) #max_new_tokens = 800,
    answer = tokenizer.batch_decode(outputs)
    # print(len(answers)+1, answer)
    answers.append(answer)
    write_json("answers.json", answers)


```
