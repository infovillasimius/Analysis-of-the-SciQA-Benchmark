```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

model, tokenizer = FastLanguageModel.from_pretrained(
    # model_name = "Llama_3.1-8B_10_epochs_DBLP",
    # model_name = "Llama_3.1-8B_10_epochs",
    # model_name = "Llama_3.1-8B_10_epochs_DBLP_SCIQA",
    model_name = "Llama_3.1-8B_10_epochs_SCIQA_with_entities",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
FastLanguageModel.for_inference(model)
```

```python
def clean(st):
    st = st.replace("\n", " ")
    st = st.replace("?", " ?")
    st = st.replace("{", " { ")
    st = st.replace("}", " } ")
    st = st.replace("\\'", "'")

    while "  " in st:
        st = st.replace("  ", " ")
    return st.strip()

def get_entities(query):
    query = clean(query)
    entities = []
    relations = []
    all_good = []
    words = query.split(" ")
    for word in words:
        if word.startswith("orkg"):
            all_good.append(word)

    for word in all_good:
        if word.startswith("orkgp:"):
            relations.append(word)
        else:
            entities.append(word)

    return {"entities": entities, "relations": relations}

from datasets import load_dataset, Dataset
# raw_dataset = load_dataset("awalesushil/DBLP-QuAD", split = "test")
raw_dataset = load_dataset("orkg/SciQA", split = "test")
questions = []
for sample in raw_dataset:
    ents = get_entities(sample["query"]["sparql"])

    element = {
        "question": sample["question"]["string"],
        "entities": ents["entities"],
        # "entities": sample["entities"],
        # "relations": sample["relations"],
        "relations": ents["relations"],
        "sparql": sample["query"]["sparql"]
    }
    questions.append(element)
```

```python
def clean(st):
    st = st.replace("\n", " ")
    st = st.replace("?", " ?")
    st = st.replace("{", " { ")
    st = st.replace("}", " } ")
    st = st.replace("\\'", "'")

    while "  " in st:
        st = st.replace("  ", " ")
    return st.strip()

def write_json(file__name, content):
    with open(file__name, "w", encoding="utf-8") as text_file:
        print(json.dumps(content, indent=4), file=text_file)


dblp_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

answers = []
for sample in tqdm(questions):
    inputs = tokenizer(
    [
        dblp_prompt.format(
            # "Given an English text describing an information need, generate a SPARQL query to retrieve relevant data from the DBLP Knowledge Graph using " + json.dumps(sample["entities"]) + " as entities and " + json.dumps(sample["relations"]) + " as relations", # instruction
            #"Given an English text describing an information need, generate a SPARQL query to retrieve relevant data from the SciQA Knowledge Graph",
            "Given an English text describing an information need, generate a SPARQL query to retrieve relevant data from the SciQA Knowledge Graph using " + json.dumps(sample["entities"]) + " as entities and " + json.dumps(sample["relations"]) + " as relations", # instruction
            sample.get("question"), # input
            "", # output - leave this blank for generation!
        )
    ], return_tensors = "pt").to("cuda")

    outputs = model.generate(**inputs, max_new_tokens = 800,   use_cache = True) #max_new_tokens = 800,
    generated_answer = tokenizer.batch_decode(outputs)
    # print(len(answers)+1, answer)
    answer = {
        "question": sample["question"],
        "sparql": clean(sample["sparql"]),
        "generated_sparql": generated_answer[0],
        # "entities": sample["entities"],
        #"relations": sample["relations"]
    }
    answers.append(answer)
    write_json("SCIQA_with_entities_model_SCIQA_with_entities.json", answers)


```
